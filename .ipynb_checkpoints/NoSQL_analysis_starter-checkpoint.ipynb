### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
dirty_establishments = establishments.find({"scores.Hygiene": 20})

# Use count_documents to display the number of documents in the result
dirty_establishments_count = establishments.count_documents({"scores.Hygiene": 20})
print(f"There are {dirty_establishments_count} establishments with a hygiene score of 20.")

# Display the first document in the results using pprint
pprint(dirty_establishments[0])
    
```

```python
# Convert the result to a Pandas DataFrame
dirty_establishments_df = pd.json_normalize(dirty_establishments)

# Display the number of rows in the DataFrame
dirty_establishments_df_rows = dirty_establishments_df.shape[0]
print(f"There are {dirty_establishments_df_rows} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
dirty_establishments_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {"LocalAuthorityName": 
                    {"$regex": "London"},
                "RatingValue":
                    {"$gte": 4}
                }
london_results = establishments.find(london_query)

# Use count_documents to display the number of documents in the result
london_query_count = establishments.count_documents(london_query)
print(f"There are {london_query_count} establishments with London as the Local Authority \
and a RatingValue greater than or equal to 4.")

# Display the first document in the results using pprint
pprint(london_results[0])
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.json_normalize(london_results)


# Display the number of rows in the DataFrame
london_df_rows = london_df.shape[0]
print(f"There are {london_df_rows} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
penang = establishments.find_one({"BusinessName": "Penang Flavours"})
penang
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
from bson.decimal128 import Decimal128, create_decimal128_context
import decimal

context = create_decimal128_context()
with decimal.localcontext(context):

    degree_search = Decimal128("0.01")
    latitude = penang["geocode"]["latitude"]
    longitude = penang["geocode"]["longitude"]
    
    min_lat = Decimal128(latitude.to_decimal() - degree_search.to_decimal())
    max_lat = Decimal128(latitude.to_decimal() + degree_search.to_decimal())
    min_lon = Decimal128(longitude.to_decimal() - degree_search.to_decimal())
    max_lon = Decimal128(longitude.to_decimal() + degree_search.to_decimal())

query = {"geocode.latitude":
            {"$gte": min_lat,
            "$lte": max_lat},
        "geocode.longitude":
            {"$gte": min_lon,
            "$lte": max_lon},
        "RatingValue": 5    
}
sort =  [("scores.Hygiene", -1)]
limit = 5

# Print the results
near_Penang = list(establishments.find(query).sort(sort).limit(limit))
pprint(near_Penang)
```

```python
# Convert result to Pandas DataFrame
near_Penang_df = pd.json_normalize(near_Penang)
near_Penang_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

match = {
    "$match": {
        "scores.Hygiene": 0
    }
}


group = {
    "$group": {
        "_id": "$LocalAuthorityName",
        "count": {
            "$sum": 1
        }
    }
}


sort = {
    "$sort": {
        "count": -1
    }
}

agg_pipeline = [match, group, sort]

# Print the number of documents in the result
results = list(establishments.aggregate(agg_pipeline))
print(f"There are {len(results)} documents in this result.")

# Print the first 10 results
pprint(results[:10])
```

```python
# Convert the result to a Pandas DataFrame
results_df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
results_df_rows = results_df.shape[0]
print(f"There are {results_df_rows} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
results_df.head(10)
```
